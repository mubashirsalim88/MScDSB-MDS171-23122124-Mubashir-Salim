The below pie chart is to understand the ratio of missed information among the students. So that when performing analysis we focus on this as a very mportant conclusion.

```python
yes=0
no=0
NotSure=0
for i in file['Missed Info']:
    if i=="Yes":
        yes=yes+1
    elif i=="No":
        no=no+1
    else:
        NotSure=NotSure+1
print("Yes:",yes,"No:",no,"Not Sure:",NotSure)
```

```python
Misinfo=[yes,no,NotSure]
labels=['Yes','No','Not Sure']
plt.title('Missied Info')
plt.pie(Misinfo, labels=labels, autopct='%1.1f%%')
```

This pie chart is an indicator to whether the information released to the student is clear or not and to what percentage are the responses collected

```python
yes=0
no=0
maybe=0
for i in file['Clarity']:
    if i=="Yes":
        yes=yes+1
    elif i=="No":
        no=no+1
    else:
        maybe=maybe+1
print("Yes:",yes,"No:",no,"Maybe:",maybe)
```

```python
Clrt=[yes,no,maybe]
labels=['Yes','No','Maybe']
plt.title('Clarity')
plt.pie(Clrt, labels=labels, autopct='%1.1f%%')
```

A heatmap is a graphical representation of data where values in a matrix are represented as colors. It is a powerful tool for visualizing complex data, especially when you want to identify patterns, correlations, or variations in a dataset. This is to find the co relation betwenn the x axis and the y axis determinants.

```python

def heatmap(df, x_colname, y_colname, figscale=1, mpl_palette_name='viridis'):
  plt.subplots(figsize=(8 * figscale, 8 * figscale))
  df_2dhist = pd.DataFrame({
      x_label: grp[y_colname].value_counts()
      for x_label, grp in df.groupby(x_colname)
  })
  sb.heatmap(df_2dhist, cmap=mpl_palette_name)
  plt.xlabel(x_colname)
  plt.ylabel(y_colname)
  return

chart = heatmap(file, *['Course', 'Missed Info'], **{})
chart
```

This heat map is based on the information clarity and where our dataset responses which was Yes , No, Maybe. The heat map helps us understand the most saturated results in the data.
